```python
import pandas as pd  # data manipulation
import numpy as np   # numerical computing
import matplotlib.pyplot as plt  # plotting
import seaborn as sns  # enhanced visualization
from sklearn.cluster import KMeans  # clustering algorithm
from sklearn.preprocessing import StandardScaler  # feature scaling
```

```python
# Load dataset from compressed CSV file
customer_data = pd.read_csv('archive.zip')  # or use a more specific name based on your data

# Display first 5 rows of the dataframe
customer_data.head()
```

```python
# Display dataset structure and summary statistics
customer_data.info()  # Shows column data types and non-null counts

# Generate descriptive statistics for numerical columns
customer_data.describe()  # Count, mean, std, min, max, quartiles

# Create pairplot visualization of variable relationships
sns.pairplot(customer_data)
plt.title("Pairwise Relationships in Customer Data")  # Add descriptive title
plt.tight_layout()  # Prevent label overlapping
plt.show()
```

```python
# Select relevant features for clustering
customer_features = customer_data[['Annual Income (k$)', 'Spending Score (1-100)']]

# Initialize and fit the feature scaler
feature_scaler = StandardScaler()

# Standardize the features (mean=0, std=1)
scaled_features = feature_scaler.fit_transform(customer_features)
```

```python
# Calculate Within-Cluster Sum of Squares (WCSS) for different cluster counts
wcss_scores = []  # Renamed from wcss to be more descriptive

for num_clusters in range(1, 11):  # More descriptive loop variable
    # Initialize and fit KMeans for each cluster count
    cluster_model = KMeans(n_clusters=num_clusters, random_state=42)
    cluster_model.fit(scaled_features)  # Using our previously scaled features

    # Store the inertia (WCSS) for each model
    wcss_scores.append(cluster_model.inertia_)

# Visualize the Elbow Method results
plt.figure(figsize=(8, 5))  # Set appropriate figure size
plt.plot(range(1, 11), wcss_scores, marker='o', linestyle='--')  # Added markers and dashed line
plt.title('Elbow Method: Optimal Number of Clusters', pad=20)  # Added padding
plt.xlabel('Number of Clusters', labelpad=10)
plt.ylabel('Within-Cluster Sum of Squares (WCSS)', labelpad=10)
plt.xticks(range(1, 11))  # Ensure all cluster numbers are shown
plt.grid(True, alpha=0.3)  # Add subtle grid
plt.tight_layout()  # Prevent label cutoff
plt.show()
```

```python
# Initialize KMeans with optimal cluster count (from elbow method)
customer_clusterer = KMeans(n_clusters=5, random_state=42)

# Assign cluster labels to each customer
cluster_labels = customer_clusterer.fit_predict(scaled_features)

# Add cluster assignments back to original dataframe
customer_data['Customer_Segment'] = cluster_labels  # More descriptive column name
```

```python
# Set up the visualization canvas
plt.figure(figsize=(10, 7), dpi=100)  # Higher resolution and size

# Create the scatter plot with enhanced styling
scatter = sns.scatterplot(
    x='Annual Income (k$)',
    y='Spending Score (1-100)',
    hue='Customer_Segment',  # Using our improved column name
    data=customer_data,
    palette='viridis',
    s=120,  # Slightly larger points
    alpha=0.8,  # Slight transparency
    edgecolor='w',  # White edges for better point distinction
    linewidth=0.5  # Edge line width
)

# Add professional styling
plt.title('Customer Segmentation by Income and Spending\nK-Means Clustering Results',
          fontsize=14, pad=20)
plt.xlabel('Annual Income (k$)', fontsize=12, labelpad=10)
plt.ylabel('Spending Score (1-100)', fontsize=12, labelpad=10)

# Improve legend and formatting
plt.legend(
    title='Customer Segment',
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    borderaxespad=0
)

# Add grid and formatting
plt.grid(True, alpha=0.3)
plt.tight_layout()

# Add value formatting
ax = plt.gca()
ax.xaxis.set_major_formatter('${x:,.0f}')

# Add annotation for context
plt.annotate('Optimal clusters identified via Elbow Method',
             xy=(0.5, -0.15),
             xycoords='axes fraction',
             ha='center',
             fontsize=10,
             color='gray')

plt.show()
```
